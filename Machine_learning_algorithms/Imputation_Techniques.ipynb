```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/ETL_DATA_new.csv")
```

```python
df
```

```python
df.set_index('TimeStamp', inplace=True)
```

```python
df
```

Remove Unnamed Columns

```python
swi=[i for i in df if 'Unnamed' in i]
df.drop(columns=swi, inplace=True,axis=1)
```

```python
df
```

```python
df=df.drop(df.columns[0], axis=1)
```

remove the columns which has 80% Nan Values

```python
nan_threshold = 0.8 * len(df)
df = df.dropna(axis=1, thresh=nan_threshold)
```

```python
df
```

# Impute the Nan values with zero

```python
df_zero = df.fillna(0)
```

```python
df_zero
```

# Impute the Nan values with Mean Method

```python
df_mean=df.fillna(df.mean())
df_mean
```

# Impute the Nan values with Median Method

```python
df_median=df.fillna(df.median())
df_median
```

# Impute the Nan values with Mode Method

```python
df_mode=df.fillna(df.mode().iloc[0])
df_mode
```

# Impute the Nan values with forward filling Method

```python
df_forward=df.fillna(method='ffill')
df_forward=df_forward.fillna(method='bfill')
df_forward
```

# Impute the Nan values with Backward filling Method

```python
df_backward=df.fillna(method='bfill')
df_backward=df_backward.fillna(method='ffill')
df_backward
```

# Impute the Nan values with KNN Method

```python
from sklearn.impute import KNNImputer
```

```python
knn_imputer = KNNImputer()
df_knn = knn_imputer.fit_transform(df)
df_knn = pd.DataFrame(df_knn, columns=df.columns)
df_knn
```
